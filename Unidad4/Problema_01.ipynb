### **Problema 1: Pesos de 68-72**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Parámetros de la distribución normal
media = 70       # Media
desv_std = 4     # Desviación estándar
total_estudiantes = 500

# VARIABLES: límites inferior y superior del intervalo
peso_lim_inf = 68
peso_lim_sup = 72

# Probabilidad acumulada hasta cada límite (CDF)
prob_menos_68 = norm.cdf(peso_lim_inf, media, desv_std)
prob_menos_72 = norm.cdf(peso_lim_sup, media, desv_std)

# Probabilidad entre 68 y 72 kg = diferencia de las dos CDFs
prob_entre_68_72 = prob_menos_72 - prob_menos_68

# Número de estudiantes en ese rango
estudiantes_entre_68_72 = total_estudiantes * prob_entre_68_72

print("Probabilidad de que un estudiante pese entre 68 y 72 kg:", prob_entre_68_72)
print("Número de estudiantes que pesan entre 68 y 72 kg:", estudiantes_entre_68_72)
```

```python
# Gráfica
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)

# Sombrear el área ENTRE los dos límites
plt.fill_between(x, y,
                 where=(x >= peso_lim_inf) & (x <= peso_lim_sup),
                 color='skyblue', alpha=0.5)

plt.title('Probabilidad de que un estudiante pese entre 68 y 72 kg.\n'
          r'$\mu=70$, $\sigma=4$, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```
