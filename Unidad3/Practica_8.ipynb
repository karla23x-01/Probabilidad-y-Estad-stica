```python
import numpy as np

X  = np.array([0, 1, 2, 3, 4, 5])
PX = np.array([0.05, 0.10, 0.25, 0.30, 0.20, 0.10])

# 1. Verificar que sumen 1
print("1) Suma de probabilidades:", np.sum(PX))

# 2. Valor esperado
EX = np.sum(X * PX)
print("2) E(X) =", EX, "unidades")

# 3. Varianza y desviación estándar
VarX = np.sum((X**2) * PX) - EX**2
print("3) Var(X) =", VarX)
print("   σ(X)   =", round(np.sqrt(VarX), 4))

# 4. Función de distribución acumulada F(x)
FX = np.cumsum(PX)
print("\n4) F(x):")
for xi, fi in zip(X, FX):
    print(f"   F({xi}) = {fi:.2f}")

# 5. Interpretación
print("\n5) Interpretación:")
print(f"   Se esperan vender {EX} unidades por semana en promedio.")
sigma = np.sqrt(VarX)
if sigma < 1.5:
    print(f"   σ = {round(sigma,4)} → demanda relativamente estable.")
else:
    print(f"   σ = {round(sigma,4)} → demanda variable.")
```
