### **Programa 4: Función de distribución acumulada (F.D.A.)**

```python
import math
import numpy as np
import matplotlib.pyplot as plt
x=[5,6,7,8,9,10]
p=[0.05,0.10,0.20,0.30,0.25,0.10]

F=np.cumsum(p)
for xi, fi in zip(x,F):
  print(f"F({xi})={fi:.2f}")
plt.step(x, F, where='post')
plt.title("funcion de distribucion acumulada (fda)")
plt.xlabel("calificacion")
plt.ylabel("probabilidad acumulada")
plt.grid(True)
plt.ylim(0,1.05)
plt.show()
```
