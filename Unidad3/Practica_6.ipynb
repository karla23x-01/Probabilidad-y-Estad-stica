```python
import numpy as np

X  = np.array([0, 1, 2, 3, 4, 5])
PX = np.array([0.10, 0.25, 0.30, 0.20, 0.10, 0.05])

# 1. Verificar que sumen 1
print("1) Suma de probabilidades:", np.sum(PX))

# 2. Valor esperado
EX = np.sum(X * PX)
print("2) E(X) =", EX)

# 3. Varianza y desviación estándar
VarX = np.sum((X**2) * PX) - EX**2
print("3) Var(X) =", VarX)
print("   σ(X)   =", np.sqrt(VarX))

# 4. Función de distribución acumulada F(x)
FX = np.cumsum(PX)
print("\n4) F(x):")
for xi, fi in zip(X, FX):
    print(f"   F({xi}) = {fi:.2f}")
```
