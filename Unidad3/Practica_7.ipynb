```python
import numpy as np

X  = np.array([0, 10, 20, 50])
PX = np.array([0.50, 0.30, 0.15, 0.05])

# 1. Verificar que sumen 1
print("1) Suma de probabilidades:", np.sum(PX))

# 2. Valor esperado
EX = np.sum(X * PX)
print("2) E(X) =", EX, "pesos")

# 3. Varianza y desviación estándar
VarX = np.sum((X**2) * PX) - EX**2
print("3) Var(X) =", VarX)
print("   σ(X)   =", round(np.sqrt(VarX), 4))

# 4. Función de distribución acumulada F(x)
FX = np.cumsum(PX)
print("\n4) F(x):")
for xi, fi in zip(X, FX):
    print(f"   F({xi}) = {fi:.2f}")

# 5. Interpretación
print("\n5) Interpretación:")
print(f"   E(X) = {EX} pesos → ganancia promedio esperada por boleto.")
print(f"   Si el boleto cuesta más de {EX} pesos, el juego NO es favorable para el jugador.")
print(f"   Si el boleto cuesta menos de {EX} pesos, el juego SÍ es favorable.")
```
