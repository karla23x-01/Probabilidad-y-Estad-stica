#Ejemplo 1

```python
#Lista de calificaciones del grupo a de probabilidad y estadística
grupoa=[70,70,65,100,95,45,88,73,28,100,96]
print(grupoa)
type(grupoa)
```

```python
#La función type nos devuelve el tipo de dato de la variable
type(grupoa)
```

```python
#Sacar la media de las calificaciones con una formula
promedio=sum(grupoa)/len(grupoa)
print(promedio)
```

```python
from statistics import mean
#Utilizar la funcion mean que esta en la libreria statistics para sacar el promedio
promedio2=mean(grupoa)
print(promedio2)
```

```python
#Determinar el valor de la mediana con la función median
mediana=median(grupoa)
print(mediana)
```

```python
#Determinar el valor de moda con la función mode
moda=mode(grupoa)
print(moda)
```

```python
#Más de una moda la función multimode
#Múltiples modas
moda2=multimode(grupoa)
print(moda2)
```

```python
#Tiempos que tardaron los 10 alumnos para los 100 metros
tiempo100mts=[15.10,17.20,14.69,13.27,22.15,18.71,19.15,20.65,15.10,17.20]
print(tiempo100mts)
```

```python
#Mean el tiempo promedio
print("El tiempo promedio de los corredores es =", mean(tiempo100mts))
print("La mediana de los tiempo de los corredores es =", median(tiempo100mts))
print("La moda de los tiempo de los corredores es =", mode(tiempo100mts))
```

```python
#LIbreria random
import random
```

```python
fabrica1=[]
for i in range (100):
  numero_aleatorio=round(random.uniform(50,100),2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
fabrica2=[]
for i in range (100):
  numero_aleatorio=round(random.uniform(70,100),2)
  fabrica2.append(numero_aleatorio)
print(fabrica2)
```

```python
#Libreria para graficar valores
import matplotlib.pyplot as plt
```

```python
plt.hist(fabrica1)
plt.title("Produción diaria de la Fabrica 1")
```

```python
plt.hist(fabrica2)
plt.title("Produción diaria de la Fabrica 2")
```

```python
media1=mean(fabrica1)
mediana1=median(fabrica1)
moda1=mode(fabrica1)

print("Valores de la fabrica 1")
print("La media es: ",media1)
print("La mediana es: ",mediana1)
print("La moda es: ",moda1)
```

```python
media2=mean(fabrica2)
mediana2=median(fabrica2)
moda2=mode(fabrica2)

print("Valores de la fabrica 2")
print("La media es: ",media2)
print("La mediana es: ",mediana2)
print("La moda es: ",moda2)
```

```python
#Generar 2 subplot
fig, (ax1, ax2)=plt.subplots(nrows=1,ncols=2,figsize=(10,5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()

```
