```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot

autos = pd.read_csv("autos_corregidos.csv", encoding="ISO-8859-1")
autos.head()
```

```python
autos.info()
```

```python
print("Valor mínimo de precio:", autos["price"].min())
print("Valor máximo de precio:", autos["price"].max())
print("Moda de precio:", autos["price"].mode()[0])
print("Media de precio:", autos["price"].mean())
print("Mediana de precio:", autos["price"].median())
print("Desviación estándar de precio:", autos["price"].std())
```

```python
autos["price"].describe()
```

```python
import math

n = len(autos["price"].dropna())
bins_sturges = math.ceil(1 + 3.322 * math.log10(n))
print(f"Cantidad óptima de bins según Sturges: {bins_sturges}")

pyplot.figure(figsize=(10, 6))
pyplot.hist(autos["price"].dropna(), bins=bins_sturges, color="steelblue", edgecolor="black")
pyplot.title("Histograma de Precios de Automóviles")
pyplot.xlabel("Precio (USD)")
pyplot.ylabel("Frecuencia")
pyplot.tight_layout()
pyplot.show()
```

```python
pyplot.figure(figsize=(12, 6))
sns.boxplot(x="body-style", y="price", data=autos)
pyplot.title("Relación entre Precio y Tipo de Carrocería")
pyplot.xlabel("Tipo de Carrocería (Body Style)")
pyplot.ylabel("Precio (USD)")
pyplot.tight_layout()
pyplot.show()
```

```python
pyplot.figure(figsize=(10, 6))
sns.boxplot(x="drive-wheels", y="price", data=autos)
pyplot.title("Relación entre Precio y Tipo de Tracción")
pyplot.xlabel("Tipo de Tracción (Drive Wheels)")
pyplot.ylabel("Precio (USD)")
pyplot.tight_layout()
pyplot.show()
```

```python
pyplot.figure(figsize=(8, 6))
sns.boxplot(x="engine-location", y="price", data=autos)
pyplot.title("Relación entre Precio y Ubicación del Motor")
pyplot.xlabel("Ubicación del Motor (Engine Location)")
pyplot.ylabel("Precio (USD)")
pyplot.tight_layout()
pyplot.show()
```


CONCLUSIONES GENERALES:

1. La mayoría de los autos tienen precios bajos o medios, pero hay algunos
   muy caros que se salen del rango normal (valores atípicos).

2. La forma del auto afecta su precio: los convertibles son los más caros
   y los hatchbacks (compactos) son los más baratos.

3. Los autos con tracción trasera cuestan más que los de tracción delantera,
   porque generalmente son autos deportivos o de lujo.

4. Los autos con el motor atrás son mucho más caros que los que tienen
   el motor adelante, ya que suelen ser autos deportivos de alto rendimiento.

5. En resumen, las características del auto como su forma, tipo de tracción
   y ubicación del motor influyen directamente en su precio. A mayor
   rendimiento y lujo, mayor es el costo del vehículo.
