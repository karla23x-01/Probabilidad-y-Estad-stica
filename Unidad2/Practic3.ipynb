```python
import math as mt
from itertools import permutations

# asistentes disponibles
asistentes = ["A1","A2","A3","A4","A5","A6","A7","A8","A9","A10"]

n = len(asistentes)
r = 4   # preguntas del examen

# fórmula de permutaciones
Pr = mt.factorial(n) / mt.factorial(n-r)

print("Número de formas de elegir asistentes:", int(Pr))

print("\nEjemplos de asignaciones (pregunta1, pregunta2, pregunta3, pregunta4):")

permutaciones_asistentes = list(permutations(asistentes, r))

# mostrar solo las primeras 10 para no imprimir 5040
for p in permutaciones_asistentes[:10]:
    print(p)
```
