```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import statsmodels.api as sm
import warnings
warnings.filterwarnings('ignore')
```

```python
base = 'https://raw.githubusercontent.com/karla23x-01/Probabilidad-y-Estad-stica/main/Unidad5/'
df1 = pd.read_csv(base + 'grupo1.csv')
df2 = pd.read_csv(base + 'grupo2.csv')
df3 = pd.read_csv(base + 'grupo3.csv')
```

```python
grupos = {'Grupo 1': df1, 'Grupo 2': df2, 'Grupo 3': df3}
for nombre, df in grupos.items():
    print(f'\n--- {nombre} ---')
    print(df.head(50))
    print(df.describe().round(2))
```

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 4))
fig.suptitle('Histogramas de Calificaciones', fontsize=13)

for ax, (nombre, df) in zip(axes, grupos.items()):
    ax.hist(df['Calificacion'], bins=10, color='steelblue', edgecolor='white')
    ax.axvline(df['Calificacion'].mean(), color='red', linestyle='--',
               label=f"Media = {df['Calificacion'].mean():.1f}")
    ax.set_title(nombre)
    ax.set_xlabel('Calificacion')
    ax.set_ylabel('Frecuencia')
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
for nombre, df in grupos.items():
    X = df[['HorasEstudio']].values
    y = df['Calificacion'].values

    modelo = LinearRegression().fit(X, y)
    pendiente  = modelo.coef_[0]
    intercepto = modelo.intercept_
    y_pred     = modelo.predict(X)
    r2         = r2_score(y, y_pred)
    r, p       = stats.pearsonr(df['HorasEstudio'], df['Calificacion'])

    print(f'\n--- {nombre} ---')
    print(f'Ecuacion:      y = {pendiente:.4f}x + {intercepto:.4f}')
    print(f'r de Pearson:  {r:.4f}')
    print(f'R2:            {r2:.4f}')
    print(f'p-valor:       {p:.4e}')
```

```python
for nombre, df in grupos.items():
    X_sm     = sm.add_constant(df['HorasEstudio'])
    modelo_sm = sm.OLS(df['Calificacion'], X_sm).fit()
    print(f'\n--- {nombre} ---')
    print(modelo_sm.summary())
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Dispersion y Regresion Lineal — Horas de Estudio vs Calificacion', fontsize=13)

for ax, (nombre, df) in zip(axes, grupos.items()):
    X = df['HorasEstudio'].values
    y = df['Calificacion'].values

    modelo    = LinearRegression().fit(X.reshape(-1, 1), y)
    pendiente = modelo.coef_[0]
    intercepto = modelo.intercept_
    r, _      = stats.pearsonr(X, y)
    r2        = r2_score(y, modelo.predict(X.reshape(-1, 1)))

    x_line = np.linspace(X.min(), X.max(), 200)
    y_line = pendiente * x_line + intercepto

    ax.scatter(X, y, color='steelblue', alpha=0.7, s=40, label='Datos')
    ax.plot(x_line, y_line, color='red', linewidth=1.5, label='Regresion')
    ax.set_title(nombre)
    ax.set_xlabel('Horas de Estudio')
    ax.set_ylabel('Calificacion Final')
    ax.text(0.05, 0.95,
            f'y = {pendiente:.3f}x + {intercepto:.2f}\nr = {r:.4f}   R2 = {r2:.4f}',
            transform=ax.transAxes, fontsize=9, verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='white', alpha=0.7))
    ax.legend()

plt.tight_layout()
plt.show()
```
